```python
import pandas as pd    # load & handle dataset
import torch           # required for running gpt2 & miniLM
from transformers import GPT2LMHeadModel, GPT2Tokenizer       # calling gpt2 model
from sentence_transformers import SentenceTransformer, util   # synonyms  similarity  , util for similarity calculation
import numpy as np     # numeric operation
import time            # measure processing time
import transformers    # load gpt2
```

```python
data = pd.read_csv('CustomerServiceDataSet.csv', encoding='latin-1')
data.head()
```

```python
EMBED_MODEL_NAME = "all-MiniLM-L6-v2" # use to find similar problem from dataset
GPT2_MODEL = "distilgpt2" # using distilgpt2 as your model
TOP_K = 3 # fetch top 3 answers from dataset
CLARITY_THRESHOLD = 0.40 # below this, ask clarification
ACCEPT_SIM_THRESHOLD = 0.65 # if semantic sim >= this, auto-accept usually
DEVICE = "cpu"
```

```python
transformers.logging.set_verbosity_error() # hige model loading warnings

embed_model = SentenceTransformer(EMBED_MODEL_NAME, device=DEVICE)    # text into vector for similarity search
problem_texts = data["problem"].astype(str).tolist()                  # extract problems from dataset
problem_embeddings = embed_model.encode(problem_texts, convert_to_tensor=True, show_progress_bar=True)    #convert problem into vector and store them into memory
```

```python
# convert text into number(tokens) for gpt2 model can understand
tokenizer = GPT2Tokenizer.from_pretrained(GPT2_MODEL)
model = GPT2LMHeadModel.from_pretrained(GPT2_MODEL).to(DEVICE)
model.config.pad_token_id = model.config.eos_token_id
```

```python
# ---------- Simple synonyms memory ----------
conversation_memory = []

#convert text into vectors and store them in a list
def mem_add(role, text):
    emb = embed_model.encode(text, convert_to_tensor=True)
    conversation_memory.append({"role": role, "text": text, "embedding": emb, "time": time.time()})

# Finds the most similar past messages using vector similarity.
def mem_retrieve_similar(query, top_k=3):
    if not conversation_memory:
        return []
    q_emb = embed_model.encode(query, convert_to_tensor=True)
    mem_embs = torch.stack([m["embedding"] for m in conversation_memory]).squeeze(1)
    sims = util.cos_sim(q_emb, mem_embs)[0]         # using consine for similarity
    topk = torch.topk(sims, k=min(top_k, len(sims)))
    results = []
    for score, idx in zip(topk.values.tolist(), topk.indices.tolist()):
        results.append({"score": float(score), "entry": conversation_memory[idx]})
    return results
```

```python
# ---------- Retrieval helper ----------

# find best matching ans from dataset on user msg
def retrieve_candidates(user_input, k=TOP_K):

  # Step 1: Check for exact keywords
    user_words = set(user_input.lower().split())
    exact_hits = []

    for idx, row in data.iterrows():
        problem_words = set(str(row["problem"]).lower().split())
        overlap = len(user_words & problem_words)
        if overlap >= 2:  # at least 2 words match exactly
            exact_hits.append({
                "score": 1.0,  # treat exact match as highest confidence
                "idx": idx,
                "problem": str(row["problem"]),
                "solution": str(row["solution"])
            })

    if exact_hits:
        # Return top k exact matches only
        return exact_hits[:k]

    # --- Step 2: If no keywords match do semantic similarity ---
    q_emb = embed_model.encode(user_input, convert_to_tensor=True)
    cos_scores = util.cos_sim(q_emb, problem_embeddings)[0]     # Finds which problems are semantically closest in meaning
    topk = torch.topk(cos_scores, k=k)
    candidates = []
    for score, idx in zip(topk.values.tolist(), topk.indices.tolist()):
        sol = str(data["solution"].iloc[idx])
        prob = str(data["problem"].iloc[idx])
        candidates.append({
            "score": float(score),
            "idx": int(idx),
            "problem": prob,
            "solution": sol
        })
    return candidates
```

```python
# ---------- GPT-2 helpers ----------

# This function takes a prompt and generates a continuation of that text using the GPT-2 model.
def gpt2_generate_text(prompt, max_new_tokens=60, stop_tokens=None):
    inputs = tokenizer(prompt, return_tensors="pt").to(DEVICE)
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=False,
        temperature=0.2,
        top_k=0,
        top_p=1.0,
        pad_token_id=model.config.eos_token_id,
    )
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return decoded[len(prompt):].strip()
```

```python
#check if solution matches users issue say accept or reject give short reason and return every thing in a clear dictionary

def gpt2_accept_reject_prompt(user_input, candidate):
    """
    Ask GPT-2 to strongly output either 'ACCEPT' or 'REJECT' followed by a short reason.
    Note: GPT-2 is not a classifier, so combine embedding heuristics with this.
    """
    prompt = (
        "You are a terse verifier. Given a customer's issue and a proposed solution, "
        "respond in EXACT format:\n\n"
        "DECISION: <ACCEPT or REJECT>\n"
        "REASON: <one short sentence>\n\n"
        f"Customer issue: {user_input}\n\nProposed solution: {candidate}\n\n"
        "DECISION:"
    )
    raw = gpt2_generate_text(prompt, max_new_tokens=40)
    # try to parse until first newline
    out = raw.strip().splitlines()
    if not out:
        return {"decision": "REJECT", "reason": "No response from verifier."}
    first_line = out[0].upper()
    decision = "ACCEPT" if "ACCEPT" in first_line else ("REJECT" if "REJECT" in first_line else None)
    reason = ""
    if decision is None:
        # fallback: look for tokens
        if "YES" in first_line or "OK" in first_line:
            decision = "ACCEPT"
        else:
            decision = "REJECT"
    # find reason if present
    if len(out) >= 2:
        reason = out[1].strip()
    else:
        # try to parse "REASON:" from raw
        if "REASON:" in raw:
            reason = raw.split("REASON:")[-1].strip().split("\n")[0]
    return {"decision": decision, "reason": reason}
```

```python
# ---------- Agent R: Retriever + Rephraser ----------

# Retrieve best matching answers
# Rephrase them professionally using GPT-2
# Enhance response with relevant memory context
# It returns polished candidate answers for further checking by Agent V.

def agent_r_retrieve_and_rephrase(user_input, k=TOP_K, rephrase_with_gpt2=True):
    candidates = retrieve_candidates(user_input, k=k)
    # Attach a short contextual memory retrieval to increase context-awareness
    mem_hits = mem_retrieve_similar(user_input, top_k=2)
    mem_context = " ".join([h["entry"]["text"] for h in mem_hits]) if mem_hits else ""
    enriched_candidates = []
    for c in candidates:
        solution = c["solution"]
        # rephrase using GPT-2 to be more conversational
        if rephrase_with_gpt2 and c["score"] >= CLARITY_THRESHOLD:
            prompt = (
                "Rephrase this solution clearly and professionally, keep meaning exactly, "
                "in one or two short sentences:\n\n"
                f"{solution}\n\nRephrased:"
            )
            try:
                rephrased = gpt2_generate_text(prompt, max_new_tokens=80)
                # simple cleanup
                if len(rephrased.split()) >= 3:
                    lines = re.split(r'[.!?]', rephrased)
                    cleaned = ". ".join([ln.strip() for ln in lines if ln.strip()][:2]).strip() + "."
                    solution_text = cleaned
                else:
                    solution_text = solution
            except Exception:
                solution_text = solution
        else:
            solution_text = solution

        enriched_candidates.append({
            "idx": c["idx"],
            "problem": c["problem"],
            "raw_solution": c["solution"],
            "solution": solution_text,
            "retrieval_score": c["score"],
            "mem_context": mem_context
        })
    return enriched_candidates
```

```python
# ---------- Agent V: Verifier + Planner ----------

# Agent V checks the quality of an answer using:
# semantic similarity
# rephrasing consistency
# GPT-2 verification
# weighted heuristic scoring
# Then decides whether the answer should be used or rejected.

def agent_v_verify_and_plan(user_input, candidate):
    """
    Returns a dict:
      {
        'acceptance': True/False,
        'score': float (combined confidence),
        'reason': str,
        'plan': optional list of step strings
      }
    """
    # Semantic alignment: query <-> candidate
    q_emb = embed_model.encode(user_input, convert_to_tensor=True)
    cand_emb = embed_model.encode(candidate["solution"], convert_to_tensor=True)
    sim_q_cand = float(util.cos_sim(q_emb, cand_emb)[0])

    # Semantic alignment to original dataset solution (helps detect rephrase drift)
    ds_emb = embed_model.encode(candidate["raw_solution"], convert_to_tensor=True)
    sim_cand_ds = float(util.cos_sim(cand_emb, ds_emb)[0])

    # Quick heuristic confidence
    heuristic_score = (candidate["retrieval_score"] * 0.6) + (sim_q_cand * 0.3) + (sim_cand_ds * 0.1)

    # Ask GPT-2 for an accept/reject decision (helps with nuance)
    try:
        vr = gpt2_accept_reject_prompt(user_input, candidate["solution"])
        decision = vr["decision"]
        reason = vr["reason"]
    except Exception as e:
        decision = "REJECT"
        reason = f"Verifier error: {e}"

    # Convert to boolean
    accepted_by_gpt2 = True if decision == "ACCEPT" else False

    # Combine heuristics
    combined_confidence = float(np.clip(heuristic_score * 1.0 + (0.15 if accepted_by_gpt2 else -0.15), 0.0, 1.0))

    # Formulate a short plan if complexity detected (simple heuristic: presence of keywords or multiple steps)
    plan = None
    clean_sol = candidate["solution"]
    clean_sol = "\n".join(
      ln for ln in clean_sol.split("\n")
      if "rephrased" not in ln.lower()
    )
    candidate["clean_solution"] = clean_sol.strip()


    # Decide final acceptance: either high combined_confidence, or gpt2 accepted + decent heuristic
    acceptance = False
    if combined_confidence >= ACCEPT_SIM_THRESHOLD:
        acceptance = True
    elif accepted_by_gpt2 and combined_confidence >= 0.45:
        acceptance = True
    else:
        acceptance = False

    return {
        "acceptance": acceptance,
        "score": combined_confidence,
        "reason": reason,
        "sim_q_cand": sim_q_cand,
        "sim_cand_ds": sim_cand_ds,
        "plan": plan
    }
```

```python
def add_sympathy(solution_text, user_input):
    """
    Adds a touch of sympathy/empathy to the solution without losing meaning.
    Uses GPT-2 to lightly rephrase.
    """
    prompt = (
        "You are a polite customer service agent. "
        "Add a short empathetic remark to this response without changing the solution meaning. "
        "Keep it 1 sentence max.\n\n"
        f"Customer said: {user_input}\n"
        f"Original solution: {solution_text}\n\n"
        "Empathetic reply:"
    )
    try:
        empathetic_text = gpt2_generate_text(prompt, max_new_tokens=50)
        # Just keep first sentence to avoid GPT-2 rambling
        empathetic_text = empathetic_text.split("\n")[0].strip()
        if empathetic_text:
            return empathetic_text
        else:
            return solution_text
    except Exception:
        return solution_text
```

```python
def dialogue_manager(user_input):
    mem_add("user", user_input)

    # 1) Retriever produces k candidates
    candidates = agent_r_retrieve_and_rephrase(user_input, k=TOP_K, rephrase_with_gpt2=True)

    # 2) Verifier evaluates each candidate
    evaluated = []
    for cand in candidates:
        v = agent_v_verify_and_plan(user_input, cand)
        evaluated.append({"candidate": cand, "verif": v})

    # 3) Sort by verif.score and prefer accepted
    evaluated = sorted(evaluated, key=lambda x: (x["verif"]["acceptance"], x["verif"]["score"]), reverse=True)
    best = evaluated[0]

    # If best not accepted or low confidence, ask clarification
    if not best["verif"]["acceptance"] or best["verif"]["score"] < 0.40:
        clarifying_question = "I couldn't find a confident match. Could you give one short extra detail (order #, date, or exact error message)?"
        mem_add("bot", clarifying_question)
        return {"reply": clarifying_question, "awaits_clarification": True, "candidates": evaluated}

    # Else build final reply: solution + empathy + confidence + follow-up
    cand = best["candidate"]
    ver = best["verif"]
    reply_parts = []

    # Original cleaned solution
    solution_text = cand.get("clean_solution", cand["solution"])

    # Optionally add empathy if problem contains negative sentiment keywords
    negative_keywords = ["issue", "problem", "trouble", "error", "delay", "failed", "cannot", "unable",]
    if any(word in user_input.lower() for word in negative_keywords):
      empathetic_text = add_sympathy(solution_text, user_input)
      # Combine empathy and solution
      solution_text = f"{empathetic_text} {solution_text}"

    reply_parts.append(solution_text)

    # If a plan exists, include succinctly
    if ver["plan"]:
        reply_parts.append("\nSuggested steps:\n" + "\n".join([f"{i+1}. {s}" for i, s in enumerate(ver["plan"])]))

    # Add confidence and follow-up
    reply_parts.append(f"\n(Confidence: {ver['score']:.2f}). Do you need anything else? (Yes/No)")

    final_reply = "\n".join(reply_parts)
    mem_add("bot", final_reply)
    return {"reply": final_reply, "awaits_clarification": False, "candidates": evaluated}
```

```python
# ---------- Main chat loop ----------
def run_chat():
    print("🤖 Hey this your Walmart Customer Sercive Chatbot. Type 'exit' to quit.\n")
    active = True

    while active:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            print("Bot: Thank you! Have a great day")
            break

        out = dialogue_manager(user_input)
        print("Bot:", out["reply"])

        if out.get("awaits_clarification"):
            clarification = input("You: ").strip()
            if clarification.lower() in ["exit", "quit"]:
                print("Bot: Bye!")
                break

            combined = user_input + " " + clarification
            out2 = dialogue_manager(combined)
            print("Bot:", out2["reply"])

            if not out2["awaits_clarification"]:
                result = follow_up_loop()
                if result == "END":
                    break
        else:
            result = follow_up_loop()
            if result == "END":
                break

def follow_up_loop():
    while True:
        follow = input("You: ").strip().lower()
        if follow in ["no", "n"]:
            print("Bot: Thanks for chatting! Goodbye")
            return "END"
        elif follow in ["yes", "y"]:
            print("Bot: Sure — please mention your issue.")
            return "CONTINUE"
        else:
            print("Bot: Please respond with Yes or No.")

if __name__ == "__main__":
    run_chat()

```
